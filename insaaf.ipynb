```python
import pandas as pd
df=pd.read_csv('census_2011.csv')
df1=df.copy()
```

```python
df.columns
```

```python
df1=df[[
    'State name',
    'District name',
    'Population',
    'Male',
    'Female',
    'Literate',
    'Male_Literate',
    'Female_Literate',
    'Rural_Households',
    'Urban_Households',
    'Households',
    'Age_Group_0_29',
    'Age_Group_30_49',
    'Age_Group_50',
    'Age not stated']]
```

### PROBLEM STATEMENT 2

```python
df1.rename(columns={
    'State name': 'State/UT',
    'District name': 'District',
    'Male_Literate': 'Literate_Male',
    'Female_Literate': 'Literate_Female',
    'Rural_Households': 'Households_Rural',
    'Urban_Households': 'Households_Urban',
    'Age_Group_0_29': 'Young_and_Adult',
    'Age_Group_30_49': 'Middle_Aged',
    'Age_Group_50': 'Senior_Citizen',
    'Age not stated': 'Age_Not_Stated'
}, inplace=True)
```

```python
df1
```

### PROBLEM STATEMENT 3

```python
def format_state_name(state):
    words = state.split()
    formatted_words = [word.lower() if word.lower() == 'and' else word.title() for word in words]
    return ' '.join(formatted_words)

# Apply the formatting function to the 'State/UT' column
df1['State/UT'] = df1['State/UT'].apply(format_state_name)
```

```python
df1['State/UT']
```

```python
df1.to_csv('census.csv',index=False)
```

```python
# Read the list of Telangana districts from the text file
telangana_districts_file = 'Telangana.txt'
with open(telangana_districts_file, 'r') as file:
    telangana_districts = [line.strip() for line in file]

# Rename the State/UT from "Andhra Pradesh" to "Telangana" for the specified districts
new_df.loc[new_df['District'].isin(telangana_districts), 'State/UT'] = 'Telangana'
```

Missing percentage calculation before filling the values.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df3 = pd.read_csv('census.csv')

# Calculate the percentage of missing data for each column before filling
missing_percentage_before = (df3.isnull().sum() / len(df3)) * 100

# Fill missing data based on provided hints
df3['Population'].fillna(df3['Male'] + df1['Female'], inplace=True)
df3['Literate'].fillna(df3['Literate_Male'] + df3['Literate_Female'], inplace=True)
df3['Young_and_Adult'].fillna(df3['Population'] - df3['Middle_Aged'] - df3['Senior_Citizen'] - df3['Age_Not_Stated'], inplace=True)
df3['Households'].fillna(df3['Households_Rural'] + df3['Households_Urban'], inplace=True)

# Calculate the percentage of missing data for each column after filling
missing_percentage_after = (df3.isnull().sum() / len(df3)) * 100

```

```python
print("missing_percentage_before : \n", missing_percentage_before)
print("-----------------------------------------------------------------------------------------------------------")
print("missing_percentage_after : \n", missing_percentage_after)
```

```python
df3.to_csv('new_census.csv',index=False)

# Visualize the impact of the data-filling process
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=['Before Filling', 'After Filling'], y=[missing_percentage_before.mean(), missing_percentage_after.mean()])
plt.title('Comparison of Missing Data Before and After Filling')
plt.ylabel('Average Percentage of Missing Data')
plt.show()
```

```python
final_census=pd.read_csv('new_census.csv')
final_housing=pd.read_csv('housing.csv')
merged_data = pd.merge(final_census, final_housing, on='Districts')

# Calculate the absolute values for Total Number of Dilapidated and Latrine_premise
merged_data['Absolute_Dilapidated'] = abs(merged_data['Total Number of Dilapidated'])
merged_data['Absolute_Latrine_premise'] = abs(merged_data['Latrine_premise'])

# Group by district and sum for both rural and urban areas
result = merged_data.groupby('districts').agg({
    'Absolute_Dilapidated': 'sum',
    'Absolute_Latrine_premise': 'sum'
}).reset_index()

# Display the result
print(result)
```
