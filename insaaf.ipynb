```python
import matplotlib.pyplot as plt
import seaborn as sns

df3 = pd.read_csv('census.csv')

# Calculate the percentage of missing data for each column before filling
missing_percentage_before = (df3.isnull().sum() / len(df3)) * 100

# Fill missing data based on provided hints
df3['Population'].fillna(df3['Male'] + df1['Female'], inplace=True)
df3['Literate'].fillna(df3['Literate_Male'] + df3['Literate_Female'], inplace=True)
df3['Young_and_Adult'].fillna(df3['Population'] - df3['Middle_Aged'] - df3['Senior_Citizen'] - df3['Age_Not_Stated'], inplace=True)
df3['Households'].fillna(df3['Households_Rural'] + df3['Households_Urban'], inplace=True)

# Calculate the percentage of missing data for each column after filling
missing_percentage_after = (df3.isnull().sum() / len(df3)) * 100
```

```python
print("missing_percentage_before : \n", missing_percentage_before)
print("-----------------------------------------------------------------------------------------------------------")
print("missing_percentage_after : \n", missing_percentage_after)
```

```python
df3.to_csv('new_census.csv',index=False)

# Visualize the impact of the data-filling process
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=['Before Filling', 'After Filling'], y=[missing_percentage_before.mean(), missing_percentage_after.mean()])
plt.title('Comparison of Missing Data Before and After Filling')
plt.ylabel('Average Percentage of Missing Data')
plt.show()
```

```python
def update_date_format(date_str):
    # Check if the value is NaN
    if pd.isna(date_str):
        return date_str
    
    
    # Split the date string by '.'
    parts = date_str.split('.')
    if len(parts) != 3:
        return date_str 
    
    # Rearrange the parts to form the YYYY-MM-DD format
    updated_date = parts[2] + '-' + parts[1] + '-' + parts[0]
    
    return updated_date

# Update the date format in the "Last_Updated" column
## Another way of changing format using inbuilt function 
## data['Last_Updated'] = pd.to_datetime(data['Last_Updated'], format='%d.%m.%Y').dt.strftime('%Y-%m-%d')


data['Last_Updated'] = data['Last_Updated'].apply(update_date_format)
```

```python

# Calculate the expected number of hospital beds based on WHO standards
merged_dataa['Expected_Beds'] = (merged_dataa['Population'] / 1000) * 3

# Calculate the shortfall or excess of hospital beds
merged_dataa['Bed_Shortfall'] = merged_dataa['Expected_Beds'] - merged_dataa['HospitalBeds']

# Visualize the difference between expected and available beds
plt.figure(figsize=(12, 8))
plt.bar(merged_dataa['State/UT'],merged_dataa['Bed_Shortfall'], color='skyblue')
plt.axhline(y=WHO_standard_bed_policy, color='red', linestyle='--', label='WHO Standard')
plt.xlabel('State/Union Territory')
plt.ylabel('Bed Shortfall/Excess')
plt.title('Difference Between Expected and Available Hospital Beds')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
