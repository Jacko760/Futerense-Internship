### PROBLEM STATEMENT 3

```python
def format_state_name(state):
    words = state.split()
    formatted_words = [word.lower() if word.lower() == 'and' else word.title() for word in words]
    return ' '.join(formatted_words)

# Apply the formatting function to the 'State/UT' column
df1['State/UT'] = df1['State/UT'].apply(format_state_name)
```

Missing percentage calculation before filling the values.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df3 = pd.read_csv('census.csv')

# Calculate the percentage of missing data for each column before filling
missing_percentage_before = (df3.isnull().sum() / len(df3)) * 100

# Fill missing data based on provided hints
df3['Population'].fillna(df3['Male'] + df1['Female'], inplace=True)
df3['Literate'].fillna(df3['Literate_Male'] + df3['Literate_Female'], inplace=True)
df3['Young_and_Adult'].fillna(df3['Population'] - df3['Middle_Aged'] - df3['Senior_Citizen'] - df3['Age_Not_Stated'], inplace=True)
df3['Households'].fillna(df3['Households_Rural'] + df3['Households_Urban'], inplace=True)

# Calculate the percentage of missing data for each column after filling
missing_percentage_after = (df3.isnull().sum() / len(df3)) * 100
```

```python
print("missing_percentage_before : \n", missing_percentage_before)
print("-----------------------------------------------------------------------------------------------------------")
print("missing_percentage_after : \n", missing_percentage_after)
```

```python
df3.to_csv('new_census.csv',index=False)

# Visualize the impact of the data-filling process
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=['Before Filling', 'After Filling'], y=[missing_percentage_before.mean(), missing_percentage_after.mean()])
plt.title('Comparison of Missing Data Before and After Filling')
plt.ylabel('Average Percentage of Missing Data')
plt.show()
```

```python
# Load the provided census and housing data files
census_data = pd.read_csv('new_census.csv')  # Replace with actual path to the census data file
housing_data = pd.read_csv('housing.csv')  # Replace with actual path to the housing data file

# Filtering necessary columns from housing data
housing_columns = ['District Name', 'Rural/Urban', 'Contition_T_Dilapidated', 'Residence_Good']
housing_filtered = housing_data[housing_columns]

# Renaming columns for clarity and consistency
housing_filtered.rename(columns={
    'District Name': 'District',
    'Contition_T_Dilapidated': 'Dilapidated_Percentage',
    'Residence_Good': 'Good_Residence_Percentage'
}, inplace=True)

# Splitting the housing data into rural and urban
housing_rural = housing_filtered[housing_filtered['Rural/Urban'] == 'Rural']
housing_urban = housing_filtered[housing_filtered['Rural/Urban'] == 'Urban']

# Merging the rural and urban housing data with the census data
merged_rural = pd.merge(census_data, housing_rural, on='District', how='left')
merged_urban = pd.merge(census_data, housing_urban, on='District', how='left')

# Calculating absolute values for rural data
merged_rural['Households_Rural_Dilapidated'] = (merged_rural['Dilapidated_Percentage'] / 100) * merged_rural['Households_Rural']
merged_rural['Households_Rural_Toilet_Premise'] = (merged_rural['Good_Residence_Percentage'] / 100) * merged_rural['Households_Rural']

# Calculating absolute values for urban data
merged_urban['Households_Urban_Dilapidated'] = (merged_urban['Dilapidated_Percentage'] / 100) * merged_urban['Households_Urban']
merged_urban['Households_Urban_Toilet_Premise'] = (merged_urban['Good_Residence_Percentage'] / 100) * merged_urban['Households_Urban']

# Selecting relevant columns for final DataFrame
final_columns_rural = ['District', 'Households_Rural', 'Households_Rural_Dilapidated', 'Households_Rural_Toilet_Premise']
final_columns_urban = ['District', 'Households_Urban', 'Households_Urban_Dilapidated', 'Households_Urban_Toilet_Premise']

final_rural = merged_rural[final_columns_rural]
final_urban = merged_urban[final_columns_urban]

# Merging rural and urban data
final_data = pd.merge(final_rural, final_urban, on='District', how='outer')

# Saving the processed data
final_data.to_csv('path_to_output_csv', index=False)  # Replace with the actual path where you want to save the output
```
